```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
df = pd.read_csv("C:\\Users\\avikk\\OneDrive\\Desktop\\walker2d-demo\\plots\\learning_curve_default.csv")

def plot_reward_data(df):
    # Convert the 'Value' column to numeric, coercing errors to NaN
    df['Value'] = pd.to_numeric(df['Value'], errors='coerce')

    # Drop rows with NaN values in the 'Value' column
    df = df.dropna(subset=['Value'])

    # Plot the data using seaborn
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='Step', y='Value', data=df, label='Rollout Episode Reward Mean')
    plt.title('Rollout Episode Reward Mean Over Steps')
    plt.xlabel('Step')
    plt.ylabel('Rollout Episode Reward Mean')
    plt.legend()
    plt.show()

    #save plot to a png file
    plt.savefig("C:\\Users\\avikk\\OneDrive\\Desktop\\walker2d-demo\\plots\\rollout_ep_rew_mean.jpg")

plot_reward_data(df)


```
